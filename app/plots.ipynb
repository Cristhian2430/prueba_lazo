```python

import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel("../data/Dataset.xlsx")
print(df.shape)
df.tail(3)
```

```python
fig = px.histogram(df, x="Edad", nbins=10, title="Distribución de Edades")
fig.show()
```

```python
df["Nombre_Completo"] = df["Nombres"] + " " + df["Apellidos"]

# Obtener la cantidad de alumnos únicos considerando Nombres y Apellidos
cantidad_alumnos_unicos = df["Nombre_Completo"].nunique()

# Crear un DataFrame con la frecuencia de cada cantidad de alumnos únicos
frecuencia_alumnos = df.groupby("Grado")["Nombre_Completo"].nunique().reset_index()

# Crear y mostrar el gráfico de distribución de la cantidad de alumnos únicos por grado
fig = px.bar(frecuencia_alumnos, x="Grado", y="Nombre_Completo",
             title="Distribución de la Cantidad de Alumnos Únicos por Grado")
fig.show()
```

```python
fig = px.box(df, x="Grado", y="Nota", title="Distribución de Notas por Grado")
fig.show()
```

```python
media_notas = df.groupby(["Curso", "Trimestre"])["Nota"].mean().reset_index()
fig = px.line(media_notas, x="Trimestre", y="Nota", color="Curso",
              title="Media de Notas por Curso en Cada Trimestre",
              markers=True,  # Agregar puntos en los valores de las líneas
              )
fig.show()
```

```python
# Crear el gráfico de distribución de notas por curso y trimestre
fig = px.box(df, x="Curso", y="Nota", color="Trimestre", title="Distribución de Notas por Curso y Trimestre")
fig.show()
```

```python
# Calcular los promedios de notas por curso
promedios_por_curso = df.groupby("Curso")["Nota"].mean().reset_index()

# Crear el gráfico de barras para comparar los promedios de notas por curso
fig = px.bar(promedios_por_curso, x="Curso", y="Nota", title="Comparación de Promedios de Notas por Curso")
fig.show()
```

```python
# Calcular los promedios de notas por trimestre
promedios_por_trimestre = df.groupby("Trimestre")["Nota"].mean().reset_index()

# Crear el gráfico de líneas para mostrar la evolución de las notas por trimestre
fig = px.line(promedios_por_trimestre, x="Trimestre", y="Nota", title="Evolución de las Notas por Trimestre",markers=True)
fig.show()
```

```python
# Calcular los promedios de notas por trimestre
promedios_por_trimestre = df.groupby("Trimestre")["Nota"].mean().reset_index()

# Crear el gráfico de líneas para mostrar la evolución de las notas por trimestre
fig = px.bar(promedios_por_trimestre, x="Trimestre", y="Nota", title="Evolución de las Notas por Trimestre")
fig.show()
```

```python
# Calcular los promedios de notas por profesor y curso
promedios_por_profesor_curso = df.groupby(["Nombres Profesor", "Curso"])["Nota"].mean().reset_index()

# Crear el gráfico de barras para comparar las notas por profesor y curso
fig = px.bar(promedios_por_profesor_curso, x="Nombres Profesor", y="Nota", color="Curso",
             title="Comparación de Notas por Profesor y Curso")
fig.show()
```
